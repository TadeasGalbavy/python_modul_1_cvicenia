```python
# 游댳 Funkcia 1: vypocitaj_priemerny_vek(studenti)
def vypocitaj_priemerny_vek(studenti):
    pocet_studentov = 0
    sucet_vek = 0

    priemerny_vek = 0

    for student in studenti:
        vek = student["vek"]
        sucet_vek += vek
        pocet_studentov += 1

    priemerny_vek = sucet_vek / pocet_studentov

    #priemerny vek sme zaokruhlili na 2 desatinne miesta, vysledok je u typu float
    # len upravujeme po캜et desatinn칳ch miest
    priemer = round(priemerny_vek, 2) 
    
    return pocet_studentov, sucet_vek, priemer

# 游댳 Funkcia 2: najmladsi_student(studenti)
def najmladsi_student(studenti):
    najmladsi_student = float("inf")
    meno_najmladsieho_studenta = None

    for student in studenti:

        if student["vek"] < najmladsi_student:
            najmladsi_student = student["vek"]
            meno_najmladsieho_studenta = student["meno"]
    
    return najmladsi_student, meno_najmladsieho_studenta

# 游댳 Funkcia 3: vypis_vysledky(priemer, meno)
def vypis_vysledky(pocet, sucet, priemer, vek, meno):

    print(f"Celkovo s칰 prihl치sen칤 {pocet} 코tudenti.")
    print(f"Priemern칳 vek 코tudentov je: {priemer} rokov.")
    print(f"Najmlad코칤 코tudent je chalanisko {meno}, ktor칳 m치 {vek} rokov.")

# 游댳 Funkcia 4: main()  
def main():
    studenti = [
    {"meno": "Anna", "vek": 22},
    {"meno": "Bra켿o", "vek": 25},
    {"meno": "Cyril", "vek": 20}
    ]

    pocet, sucet, priemer = vypocitaj_priemerny_vek(studenti)
    vek, meno = najmladsi_student(studenti)

    vypis_vysledky(pocet, sucet, priemer, vek, meno)

# 游댳 Spustenie hlavn칠ho programu
if __name__ == "__main__":
    main()
```
