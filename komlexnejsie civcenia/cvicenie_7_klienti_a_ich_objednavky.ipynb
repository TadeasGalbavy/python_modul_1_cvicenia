```python
# üß† √öLOHA: Evidencia objedn√°vok z√°kazn√≠kov (z√°klady d√°tovej anal√Ωzy v Pythone)
#
# Tvoj cieƒæ:
# Nap√≠≈° program, ktor√Ω bude zaznamen√°va≈• a vyhodnocova≈• √∫daje o objedn√°vkach z√°kazn√≠kov.
#
# ====================================
# 1. ZBIERANIE √öDAJOV (cez input):
#    - meno z√°kazn√≠ka
#    - n√°zov produktu
#    - poƒçet kusov (cel√© ƒç√≠slo)
#    - cena za kus (desatinn√© ƒç√≠slo)
#
# 2. ULO≈ΩENIE √öDAJOV:
#    - navrhni vhodn√∫ ≈°trukt√∫ru (zoznam, slovn√≠ky, ...), aby si vedel nesk√¥r d√°ta spracova≈•
#
# 3. V√ùSTUP (keƒè pou≈æ√≠vateƒæ ukonƒç√≠ zad√°vanie):
#    - vyp√≠≈° v≈°etky objedn√°vky (prehƒæadne)
#    - vypoƒç√≠taj a vyp√≠≈°:
#      a) celkov√∫ hodnotu v≈°etk√Ωch objedn√°vok
#      b) poƒçet objedn√°vok pre ka≈æd√©ho z√°kazn√≠ka zvl√°≈°≈•
#      c) najdrah≈°iu objedn√°vku (na z√°klade: poƒçet kusov √ó cena za kus)
#
# ====================================
# üîÅ EXTRA (nepovinn√©):
#    - umo≈æni ukonƒçi≈• zad√°vanie zadan√≠m pr√≠kazu "stop"
#    - ak pou≈æ√≠vateƒæ zad√° zl√Ω vstup (napr. text namiesto ƒç√≠sla), o≈°etri to
```

```python
objednavky_zakaznikov = [] #list objedktov kam zbierame udaje o kazdej objednavke

while True:
    #inputy
    meno_zanaknika = input("Zadaj meno zakaznika: ")
    nazov_produktu = input("Zadaj n√°zov produktu: ")
    #validacia
    while True:
        try:
            pocet_kusov = int(input("Zadaj poƒçet kusov: "))
            break
        except ValueError:
            print("Poƒçet mus√≠ by≈• cel√© ƒç√≠slo!")
    #validacia
    while True:
        try:
            cena_produktu = float(input("Zadaj cenu produktu: "))
            break
        except ValueError:
            print("Poƒçet mus√≠ by≈• cel√© ƒç√≠slo!")

    poznamka = input("Zadaj pozn√°mku: ")
    #ak je poznamka prazdna, vypise sa hodnota bez poznamky
    if poznamka == "":
        poznamka = "bez poznamky"

    #tvorba slovnika a pridavanie hodnot pre 1 cyklus (jednu objedn√°vku)
    objednavka = {}
    objednavka["Meno"] = meno_zanaknika
    objednavka["Produkt"] = nazov_produktu
    objednavka["Poƒçet kusov"] = pocet_kusov
    objednavka["Cena"] = cena_produktu
    objednavka["Poznamka"] = poznamka

    #na konci cyklu objednavku appendujeme do listu objektov
    objednavky_zakaznikov.append(objednavka)

    #pytame sa ci cyklus opakujeme alebo ukoncujeme
    nova_objednavka = input("Chce≈° zada≈• nov√∫ objedn√°vku? (ano/nie)")
    if nova_objednavka.lower() == "nie":
        break

suma_objednavok = 0 #poƒçitadlo sumy v≈°etk√Ωch objedn√°vok spolu
pocet_obj_na_zakaznika = {} #slovnik mien a poctu objedn√°vok
max_hodnota_objednavky = 0 #definujeme poƒçiatoƒçn√∫ hodnotu objednavky s najvy≈°≈°ou hodnotou
najdrahsia_objednavka = None #definujeme obj s najvy≈°≈°ou hodnotou|None preto lebo sa jedn√° o objekt kam vraciame riadok zo slovn√≠ka
naj_zakaznik = None #definujeme meno zakaznika s najvacsim poctom obj | none preto lebo sa jedna o objekt
naj_pocet = 0 #premenna pre najvyssi pocet objedn√°vok

for objednavka in objednavky_zakaznikov:
    #pocitanie hodnoty kazdej objedn√°vky
    vypocet_hodnoty_objednavky = objednavka["Poƒçet kusov"] * objednavka["Cena"]
    
    #pripocitanie sumy objednavky do premennej ktor√° zbiera celkovu honotu v≈°etkych objedn√°vok
    suma_objednavok += vypocet_hodnoty_objednavky

    #do premennej zakaznik si vytahujeme hodnotu z kluca Meno
    zakaznik = objednavka["Meno"]

    #podmienka kde zistujeme ci je zakaznik v zozname pocet_obj_na_zajaznika
    if zakaznik in pocet_obj_na_zakaznika:
        pocet_obj_na_zakaznika[zakaznik] += 1 #ak je, pripocitame jednu objednovku
    else:
        pocet_obj_na_zakaznika[zakaznik] = 1 #ak nie je, zalozime zaznam a nastavime na pocet 1

    # kontrolujeme vysky objednavok
    if vypocet_hodnoty_objednavky > max_hodnota_objednavky:
        max_hodnota_objednavky = vypocet_hodnoty_objednavky #prideniele hodnoty pri spolneni podmienky, ak pride obj s najvy≈°≈°≈°ou hodnotou
        najdrahsia_objednavka = objednavka #tu vytahujeme objednavku (doƒçasn√° premenna vo for cykle) s najvy≈°≈°ou hodnotou

# cyklus kde zistujeme pocet objednavok na klienta / kluc
# docasna premenna meno n√°m vracia n√°zov kluca
for meno in pocet_obj_na_zakaznika:
    if pocet_obj_na_zakaznika[meno] > naj_pocet: #porovnavame ci hodnota kluca slovnika je vyssia ako hodnota premennej naj_pocet
        naj_pocet = pocet_obj_na_zakaznika[meno] #vraciame hodnotu kluca v docasnej premennej meno
        naj_zakaznik = meno #vraciame nazov kluca z docasnej premennej vo for cykle
```

```python
print(objednavky_zakaznikov)
print(pocet_obj_na_zakaznika)
print(f"Celkov√° hodnota objedn√°vok: {suma_objednavok}‚Ç¨")
print(f"Najvy≈°≈°ia hodnota z objedn√°vok je: {max_hodnota_objednavky}‚Ç¨")
print(f"Jedn√° sa o: {najdrahsia_objednavka}")
print(f"Naj zakaznik je: {naj_zakaznik} s poƒçtom {naj_pocet}.")
```
