```python
# ----------------------------------------------------------------------
# ÚLOHA – Cestovná agentúra (spracovanie predajov zájazdov)
# ----------------------------------------------------------------------
# Máš zoznam objednávok, kde každá obsahuje:
# - "Zakaznik": meno zákazníka
# - "Destinacia": cieľová krajina zájazdu
# - "Cena": cena zájazdu v eurách
#
# Tvoja úloha:
# 1. Vypočítaj celkový obrat podľa zákazníka → {meno: súčet cien}
# 2. Zisti, kto minul najviac peňazí (meno + suma)
# 3. Spočítaj, koľkokrát sa predala každá destinácia → {destinacia: počet predajov}
# 4. Nájdi najmenej predávanú destináciu (názov + počet)
#
# ➕ Všetko napíš najprv ako jeden veľký skript.
# ➕ Až potom si spoločne urobíme refaktoring do 4 funkcií.
# ----------------------------------------------------------------------

zajazdy = [
    {"Zakaznik": "Eva", "Destinacia": "Taliansko", "Cena": 499},
    {"Zakaznik": "Jozef", "Destinacia": "Grécko", "Cena": 599},
    {"Zakaznik": "Eva", "Destinacia": "Taliansko", "Cena": 499},
    {"Zakaznik": "Anna", "Destinacia": "Chorvátsko", "Cena": 399},
    {"Zakaznik": "Eva", "Destinacia": "Grécko", "Cena": 599},
    {"Zakaznik": "Jozef", "Destinacia": "Taliansko", "Cena": 499},
    {"Zakaznik": "Anna", "Destinacia": "Chorvátsko", "Cena": 399}
]

obrat_podla_zakaznika = {} # 1. Vypočítaj celkový obrat podľa zákazníka → {meno: súčet cien}
top_klient = None # 2. Zisti, kto minul najviac peňazí (meno + suma)
top_suma = 0 # 2. Zisti, kto minul najviac peňazí (meno + suma)
destinacie_predaj = {} # 3. Spočítaj, koľkokrát sa predala každá destinácia → {destinacia: počet predajov}
min_destinacia = float("inf") # 4. Nájdi najmenej predávanú destináciu (názov + počet)
min_destinacia_nazov = None # 4. Nájdi najmenej predávanú destináciu (názov + počet)
zoznam_min_destinacii = [] #bonus vypíšeme všetky min destinacie, nie len prvý výskyt

for zaznam in zajazdy:
    meno_zakaznika = zaznam["Zakaznik"]
    suma_zajazdu = zaznam["Cena"]
    nazov_destinacie = zaznam["Destinacia"]

    # 1. Vypočítaj celkový obrat podľa zákazníka → {meno: súčet cien}
    if meno_zakaznika in obrat_podla_zakaznika:
        obrat_podla_zakaznika[meno_zakaznika] += suma_zajazdu
    else:
        obrat_podla_zakaznika[meno_zakaznika] = suma_zajazdu
    
    # 2. Zisti, kto minul najviac peňazí (meno + suma)
    if obrat_podla_zakaznika[meno_zakaznika] > top_suma:
        top_suma = obrat_podla_zakaznika[meno_zakaznika]
        top_klient = meno_zakaznika
    
    # 3. Spočítaj, koľkokrát sa predala každá destinácia → {destinacia: počet predajov}
    if nazov_destinacie in destinacie_predaj:
        destinacie_predaj[nazov_destinacie] += 1
    else:
        destinacie_predaj[nazov_destinacie] = 1

# 4. Nájdi najmenej predávanú destináciu (názov + počet)
for destinacia in destinacie_predaj:

    if destinacie_predaj[destinacia] < min_destinacia:
        min_destinacia = destinacie_predaj[destinacia]
        min_destinacia_nazov = destinacia

#bonus vypíšeme všetky min destinacie
for destinacia in destinacie_predaj:
    if destinacie_predaj[destinacia] == min_destinacia:
       zoznam_min_destinacii.append((destinacia, destinacie_predaj[destinacia]))


print(top_suma, top_klient) # 2. Zisti, kto minul najviac peňazí (meno + suma)
print(obrat_podla_zakaznika) # 1. Vypočítaj celkový obrat podľa zákazníka → {meno: súčet cien}
print(destinacie_predaj) # 3. Spočítaj, koľkokrát sa predala každá destinácia → {destinacia: počet predajov}
print(min_destinacia, min_destinacia_nazov) # 4. Nájdi najmenej predávanú destináciu (názov + počet)
print(zoznam_min_destinacii) #bonus vypíšeme všetky min destinacie
```

```python
# 1. Funkcia: vypocitaj_obrat_podla_zakaznika(zajazdy)
#    - vstup: zoznam zajazdov
#    - výstup: slovník {meno zákazníka: súčet jeho objednávok}

zajazdy = [
    {"Zakaznik": "Eva", "Destinacia": "Taliansko", "Cena": 499},
    {"Zakaznik": "Jozef", "Destinacia": "Grécko", "Cena": 599},
    {"Zakaznik": "Eva", "Destinacia": "Taliansko", "Cena": 499},
    {"Zakaznik": "Anna", "Destinacia": "Chorvátsko", "Cena": 399},
    {"Zakaznik": "Eva", "Destinacia": "Grécko", "Cena": 599},
    {"Zakaznik": "Jozef", "Destinacia": "Taliansko", "Cena": 499},
    {"Zakaznik": "Anna", "Destinacia": "Chorvátsko", "Cena": 399}
]

def vypocitaj_obrat_podla_zakaznika(zajazdy):
    obrat_podla_zakaznika = {}

    for zaznam in zajazdy:
        meno_zakaznika = zaznam["Zakaznik"]
        suma_zajazdu = zaznam["Cena"]

        # 1. Vypočítaj celkový obrat podľa zákazníka → {meno: súčet cien}
        if meno_zakaznika in obrat_podla_zakaznika:
            obrat_podla_zakaznika[meno_zakaznika] += suma_zajazdu
        else:
            obrat_podla_zakaznika[meno_zakaznika] = suma_zajazdu
    
    return obrat_podla_zakaznika

print(vypocitaj_obrat_podla_zakaznika(zajazdy))

# 2. Funkcia: najdi_top_klienta(obraty)
#    - vstup: slovník z funkcie č. 1 (meno → súčet)
#    - výstup: meno zákazníka s najvyššou útratou + suma (napr. tuple)

def najdi_top_klienta(obraty):
    
    top_klient = None # 2. Zisti, kto minul najviac peňazí (meno + suma)
    top_suma = 0

    for meno in obraty:

        if obraty[meno] > top_suma:
            top_suma = obraty[meno]
            top_klient = meno

    return top_klient, top_suma

obrat = vypocitaj_obrat_podla_zakaznika(zajazdy)
print(najdi_top_klienta(obrat))

# 3. Funkcia: spocitaj_predaje_podla_destinacie(zajazdy)
#    - Vstup: zoznam zajazdy
#    - Výstup: slovník {destinacia: počet predajov}
#    - Úloha: Spočítať, koľkokrát sa predala každá destinácia

def spocitaj_predaje_podla_destinacie(zajazdy):
    destinacie_predaj = {}

    for zaznam in zajazdy:
        nazov_destinacie = zaznam["Destinacia"]

        if nazov_destinacie in destinacie_predaj:
            destinacie_predaj[nazov_destinacie] += 1
        else:
            destinacie_predaj[nazov_destinacie] = 1

    return destinacie_predaj

print(spocitaj_predaje_podla_destinacie(zajazdy))

# 4. Funkcia: najdi_min_destinacie(predaje)
#    - Vstup: slovník {destinacia: počet predajov}
#    - Výstup: zoznam destinácií s najnižším predajom
#    - Úloha: Nájde všetky destinácie s rovnakým najnižším počtom predajov

def najdi_min_destinacie(predaje):

    min_destinacia = float("inf")
    min_destinacia_nazov = None
    zoznam_min_destinacii = []
    
    for destinacia in predaje:

        if predaje[destinacia] < min_destinacia:
            min_destinacia = predaje[destinacia]
            min_destinacia_nazov = destinacia

            # keď máme novú minimálnu hodnotu, vyčistíme zoznam a pridáme novú
            zoznam_min_destinacii = [(destinacia, predaje[destinacia])]

        elif predaje[destinacia] == min_destinacia:
            zoznam_min_destinacii.append((destinacia, predaje[destinacia]))
            
    return min_destinacia, min_destinacia_nazov, zoznam_min_destinacii

predaj = spocitaj_predaje_podla_destinacie(zajazdy)
print(najdi_min_destinacie(predaj))
```
