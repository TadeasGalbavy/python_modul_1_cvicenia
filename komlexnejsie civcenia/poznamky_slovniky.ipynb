### 1. ŠTRUKTÚRA

| Typ     | Vzor                    | Obsah                      |
| ------- | ----------------------- | -------------------------- |
| Zoznam  | zoznam = [10, 20, 30] | Radené prvky podľa poradia |
| Slovník | slovnik = {"Eva": 37} | Páry: **kľúč → hodnota**   |


### 2. PRECHÁDZANIE for CYKLOM

| Cieľ           | Zoznam             | Slovník                               |
| -------------- | ------------------ | --------------------------------------|
| Každý prvok    | for x in zoznam:   | for kluc in slovnik:                  |
| Kľúč a hodnota | —                  | for kluc, hodnota in slovnik.items(): |


### 3. PRÍSTUP K HODNOTE

| Cieľ               | Zoznam           | Slovník                               |
| ------------------ | ---------------- | --------------------------------------|
| Prvok podľa indexu | zoznam[0] → 10   | slovnik["Eva"] → 37                   |
| Kľúč neexistuje    | IndexError       | KeyError (alebo get() pre bezpečnosť) |

### 4. PRIDÁVANIE NOVÝCH HODNÔT

| Operácia         | Zoznam              | Slovník               |
| ---------------- | ------------------- | --------------------- |
| Pridať na koniec | zoznam.append(40)   | slovnik["Jozef"] = 50 |
| Zmeniť hodnotu   | zoznam[0] = 99      | slovnik["Eva"] = 42   |

### 5. PODMIENKA if in

| Chceš zistiť:           | Zoznam             | Slovník                |
| ----------------------- | ------------------ | ---------------------- |
| Či je hodnota v zozname | if 10 in zoznam:   | —                      |
| Či je kľúč v slovníku   | —                  | if "Eva" in slovnik:   |

### 6. PRÍKLADY

Zoznam:

zoznam = [10, 20, 30]

for cislo in zoznam:
    print(cislo)

Slovník:

slovnik = {"Eva": 37, "Jozef": 15}

for meno in slovnik:
    print(meno, slovnik[meno]) #meno vracia kľúč, slovnik[meno] vracia jeho hodnotu

### 7. rozdiel v zátvorkách

tu sú uvodzovky lebo sa pýtam na kluc meno
objednavka = {"Meno": "Eva"}
print(objednavka["Meno"])  # → Eva

tu nie sú uvodzovky lebo sa pátam na premennu do ktorej som si kluc uložil
krajina = "Slovensko"
predane_kusy = {"Slovensko": 120}

print(predane_kusy[krajina])  # → 120
