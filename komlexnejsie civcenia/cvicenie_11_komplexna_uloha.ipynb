```python
# ----------------------------------------------------------------------
# ðŸŽ¯ CIEÄ½: RozbiÅ¥ veÄ¾kÃ½ skript na 4 samostatnÃ© a znovupouÅ¾iteÄ¾nÃ© funkcie
# ----------------------------------------------------------------------
# MÃ¡Å¡ zoznam trÅ¾ieb (zoznam slovnÃ­kov), kde kaÅ¾dÃ½ zÃ¡znam obsahuje:
# - nÃ¡zov mesta
# - krajinu
# - poÄet predanÃ½ch kusov
# - cenu za kus
#
# Tento zoznam je zadanÃ½ v premennej `trzby`, ktorÃº budeÅ¡ odovzdÃ¡vaÅ¥ do funkciÃ­.
#
# ----------------------------------------------------------------------
# TVOJE FUNKCIE BUDÃš:
#
# 1. Funkcia: vypocitaj_trzby_podla_krajiny(trzby)
#    - vstup: zoznam trzby
#    - vÃ½stup: slovnÃ­k {krajina: sÃºÄet trÅ¾by (kusy Ã— cena)}
#
# 2. Funkcia: najdi_top_krajinu(trzby)
#    - vstup: zoznam trzby
#    - vÃ½stup: tuple (krajina, suma) â†’ krajina s najvyÅ¡Å¡ou trÅ¾bou
#
# 3. Funkcia: vypocitaj_kusy_podla_mesta(trzby)
#    - vstup: zoznam trzby
#    - vÃ½stup: slovnÃ­k {mesto: poÄet predanÃ½ch kusov spolu}
#
# 4. Funkcia: najdi_top_mesto(trzby)
#    - vstup: zoznam trzby
#    - vÃ½stup: tuple (mesto, poÄet kusov) â†’ mesto s najvyÅ¡Å¡Ã­m poÄtom predanÃ½ch kusov
#
# ----------------------------------------------------------------------
# POZOR:
# - Å¾iadna funkcia nesmie pouÅ¾Ã­vaÅ¥ globÃ¡lne premennÃ© (vÅ¡etko musÃ­ byÅ¥ vo vnÃºtri)
# - kaÅ¾dÃ¡ funkcia mÃ¡ jeden jasnÃ½ vÃ½stup
# - keÄ budÃº hotovÃ© vÅ¡etky, budeÅ¡ ich volaÅ¥ na konci v hlavnom skripte
# ----------------------------------------------------------------------
trzby = [
    {"Mesto": "Bratislava", "Krajina": "Slovensko", "Kusy": 20, "Cena": 9.99},
    {"Mesto": "Praha", "Krajina": "ÄŒesko", "Kusy": 25, "Cena": 8.5},
    {"Mesto": "KoÅ¡ice", "Krajina": "Slovensko", "Kusy": 15, "Cena": 9.99},
    {"Mesto": "Brno", "Krajina": "ÄŒesko", "Kusy": 10, "Cena": 8.5},
    {"Mesto": "Ostrava", "Krajina": "ÄŒesko", "Kusy": 18, "Cena": 8.5},
    {"Mesto": "PreÅ¡ov", "Krajina": "Slovensko", "Kusy": 12, "Cena": 9.99}
]

# 1. Funkcia: vypocitaj_trzby_podla_krajiny(trzby)
#    - vstup: zoznam trzby
#    - vÃ½stup: slovnÃ­k {krajina: sÃºÄet trÅ¾by (kusy Ã— cena)}
def vypocitaj_trzby_podla_krajiny(trzby):
    celkove_trzby_podla_krajiny = {}
    for krajina in trzby:

        krajina_var = krajina["Krajina"]
        sucet_trzby = krajina["Kusy"] * krajina["Cena"]
        
        if krajina_var in celkove_trzby_podla_krajiny:
            celkove_trzby_podla_krajiny[krajina_var] += sucet_trzby
        else:
            celkove_trzby_podla_krajiny[krajina_var] = sucet_trzby
            
    return celkove_trzby_podla_krajiny

vysledok = vypocitaj_trzby_podla_krajiny(trzby) #dÃ¡vame do premennej lebo tuto premennÃº mÃ´Å¾eme pouÅ¾iÅ¥ v inej funkcii

#print(vysledok)


# 2. Funkcia: najdi_top_krajinu(trzby)
#    - vstup: zoznam trzby
#    - vÃ½stup: tuple (krajina, suma) â†’ krajina s najvyÅ¡Å¡ou trÅ¾bou

def najdi_top_krajinu(trzby):
 
    top_krajina_suma = None
    top_zarobok = 0

    trzby_podla_krajiny = vypocitaj_trzby_podla_krajiny(trzby)

    for krajina in trzby_podla_krajiny:

        if trzby_podla_krajiny[krajina] > top_zarobok:
            top_zarobok = trzby_podla_krajiny[krajina]
            top_krajina_suma = krajina

    return top_krajina_suma, top_zarobok

#nedal som do premennej vysledok, je to len moÅ¾nÃ¡ varianta - toto ale nemÃ´Ã½eme pouÅ¾iÅ¥ v inej funkcii
#print(najdi_top_krajinu(trzby))

# 3. Funkcia: vypocitaj_kusy_podla_mesta(trzby)
#    - vstup: zoznam trzby
#    - vÃ½stup: slovnÃ­k {mesto: poÄet predanÃ½ch kusov spolu}

def vypocitaj_kusy_podla_mesta(trzby):
    celkove_predane_kusy = {}

    for zaznam in trzby:
        predane_kusy = zaznam["Kusy"]
        mesto = zaznam["Mesto"]
        
        if mesto in celkove_predane_kusy:
            celkove_predane_kusy[mesto] += predane_kusy
        else:
            celkove_predane_kusy[mesto] = predane_kusy
    
    return celkove_predane_kusy

vysledok = vypocitaj_kusy_podla_mesta(trzby) #dÃ¡vame do premennej lebo tuto premennÃº mÃ´Å¾eme pouÅ¾iÅ¥ v inej funkcii

#print(vysledok)


# 4. Funkcia: najdi_top_mesto(trzby)
#    - vstup: zoznam trzby
#    - vÃ½stup: tuple (mesto, poÄet kusov) â†’ mesto s najvyÅ¡Å¡Ã­m poÄtom predanÃ½ch kusov

def najdi_top_mesto(trzby):

    predaj = vypocitaj_kusy_podla_mesta(trzby)

    top_mesto = None
    top_kusy = 0

    for krajina in predaj:
        if predaj[krajina] > top_kusy:
            top_kusy = predaj[krajina]
            top_mesto = krajina

    return top_kusy, top_mesto

vysledok = najdi_top_mesto(trzby)

#print(vysledok)

print(vypocitaj_trzby_podla_krajiny(trzby))
print(najdi_top_krajinu(trzby))
print(vypocitaj_kusy_podla_mesta(trzby))
print(najdi_top_mesto(trzby))
```
