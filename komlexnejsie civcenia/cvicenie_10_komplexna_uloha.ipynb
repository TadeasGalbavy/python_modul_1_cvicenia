```python
# ---------------------------------------------------------------
# KOMPLEXNÁ ÚLOHA: ANALÝZA OBJEDNÁVOK (PRÁCA SO SLOVNÍKMI)
# ---------------------------------------------------------------
# Máš nasledujúci zoznam objednávok:
objednavky = [
    {"Meno": "Eva", "Produkt": "Tričko", "Cena": 14.9},
    {"Meno": "Jozef", "Produkt": "Hrnček", "Cena": 5.99},
    {"Meno": "Eva", "Produkt": "Mikina", "Cena": 29.9},
    {"Meno": "Anna", "Produkt": "Tričko", "Cena": 14.9},
    {"Meno": "Eva", "Produkt": "Hrnček", "Cena": 5.99},
    {"Meno": "Jozef", "Produkt": "Tričko", "Cena": 14.9},
    {"Meno": "Anna", "Produkt": "Mikina", "Cena": 29.9}
]
#
# Vypíš tieto výstupy:
# 1. Súčet objednávok pre každého zákazníka (meno → suma)
# 2. Koľko kusov každého produktu sa predalo (produkt → počet)
# 3. Kto nakúpil najviac (suma) – meno + hodnota
# 4. Ktorý produkt sa predal najčastejšie – produkt + počet
#
# Nezabudni:
# - správne inicializuj slovníky
# - pracuj so `for` cyklom a kontroluj výskyt kľúčov

objednavky = [
    {"Meno": "Eva", "Produkt": "Tričko", "Cena": 14.9},
    {"Meno": "Jozef", "Produkt": "Hrnček", "Cena": 5.99},
    {"Meno": "Eva", "Produkt": "Mikina", "Cena": 29.9},
    {"Meno": "Anna", "Produkt": "Tričko", "Cena": 14.9},
    {"Meno": "Eva", "Produkt": "Hrnček", "Cena": 5.99},
    {"Meno": "Jozef", "Produkt": "Tričko", "Cena": 14.9},
    {"Meno": "Anna", "Produkt": "Mikina", "Cena": 29.9}
]

sucet_obj = {} # 1.
predane_produkty = {} # 2.
max_suma = 0 # 3.
top_klient = None # 3.
predaj_produktov = {} # 4.
min_predaj = float("inf")
min_produkt = None

for objednavka in objednavky:
    meno = objednavka["Meno"]
    suma = objednavka["Cena"]
    produkt = objednavka["Produkt"]
    
    if meno in sucet_obj:
        sucet_obj[meno] += suma
    else:
        sucet_obj[meno] = suma
        
    if produkt in predane_produkty:
        predane_produkty[produkt] += 1
    else:
        predane_produkty[produkt] = 1

    if sucet_obj[meno] > max_suma:
        max_suma = sucet_obj[meno]
        top_klient = meno

    if produkt in predaj_produktov:
        predaj_produktov[produkt] += suma
    else:
        predaj_produktov[produkt] = suma

for produkt in predaj_produktov:

    if predaj_produktov[produkt] < min_predaj:
        min_predaj = predaj_produktov[produkt]
        min_produkt = produkt

print(sucet_obj)
print(predane_produkty)
print(max_suma, top_klient)
print(predaj_produktov)
print(min_predaj, min_produkt)
```

```python
# ------------------------------------------------------------
# REFÁKTORING ÚLOHY: ANALÝZA OBJEDNÁVOK
# ------------------------------------------------------------
# ÚLOHA:
# Máš pôvodný kód, ktorý analyzuje objednávky zákazníkov.
# Teraz rozdeľ túto analýzu do samostatných funkcií.
#
# Každá funkcia dostane ako vstup parameter `objednavky`
# a vráti svoj výsledok.
#
# ------------------------------------------------------------
# ZOZNAM FUNKCIÍ, KTORÉ MÁŠ VYTVORIŤ:
# ------------------------------------------------------------

# 1. Funkcia: vypocitaj_sucet_podla_mena(objednavky)
#    - Vráti slovník {meno → súčet všetkých objednávok}

# 2. Funkcia: spocitaj_produkty(objednavky)
#    - Vráti slovník {produkt → počet kusov predaných spolu}

# 3. Funkcia: najdi_top_klienta(sucet_obj)
#    - Vstup: slovník zo súčtov
#    - Vráti meno zákazníka s najvyšším súčtom + túto hodnotu

# 4. Funkcia: vypocitaj_predaj_podla_produkta(objednavky)
#    - Vráti slovník {produkt → celkový predaj v €}

# 5. Funkcia: najdi_najmenej_predavany(predaj_produktov)
#    - Vstup: slovník z bodu 4
#    - Vráti produkt s najnižším predajom v € + túto hodnotu

# ------------------------------------------------------------
# Na konci programu volaj všetky funkcie a vypíš výsledky.
# Všetko by malo byť čitateľné a funkčné.
# ------------------------------------------------------------

objednavky = [
    {"Meno": "Eva", "Produkt": "Tričko", "Cena": 14.9},
    {"Meno": "Jozef", "Produkt": "Hrnček", "Cena": 5.99},
    {"Meno": "Eva", "Produkt": "Mikina", "Cena": 29.9},
    {"Meno": "Anna", "Produkt": "Tričko", "Cena": 14.9},
    {"Meno": "Eva", "Produkt": "Hrnček", "Cena": 5.99},
    {"Meno": "Jozef", "Produkt": "Tričko", "Cena": 14.9},
    {"Meno": "Anna", "Produkt": "Mikina", "Cena": 29.9}
]

# 1. Funkcia: vypocitaj_sucet_podla_mena(objednavky)
#    - Vráti slovník {meno → súčet všetkých objednávok}
def vypocitaj_sucet_podla_mena(objednavky):
    sucet_obj = {}
    for objednavka in objednavky:
        meno = objednavka["Meno"]
        suma = objednavka["Cena"]
        produkt = objednavka["Produkt"]

        if meno in sucet_obj:
            sucet_obj[meno] += suma
        else:
            sucet_obj[meno] = suma
        
    return(sucet_obj)
vysledok = vypocitaj_sucet_podla_mena(objednavky)


# 2. Funkcia: spocitaj_produkty(objednavky)
#    - Vráti slovník {produkt → počet kusov predaných spolu}
def spocitaj_produkty(objednavky):
    predane_produkty = {}
    for objednavka in objednavky:
            meno = objednavka["Meno"]
            suma = objednavka["Cena"]
            produkt = objednavka["Produkt"]
            
            if produkt in predane_produkty:
                predane_produkty[produkt] += 1
            else:
                predane_produkty[produkt] = 1
        
    return(predane_produkty)

vysledok = spocitaj_produkty(objednavky)


# 3. Funkcia: najdi_top_klienta(sucet_obj)
#    - Vstup: slovník zo súčtov
#    - Vráti meno zákazníka s najvyšším súčtom + túto hodnotu
def najdi_top_klienta(sucet_obj):
    max_suma = 0
    top_klient = None

    for meno in sucet_obj:
        suma = sucet_obj[meno]

        if suma > max_suma:
            max_suma = suma
            top_klient = meno
    
    return(max_suma, top_klient)

sucet = vypocitaj_sucet_podla_mena(objednavky)

vystup = najdi_top_klienta(sucet)


# 4. Funkcia: vypocitaj_predaj_podla_produkta(objednavky)
#    - Vráti slovník {produkt → celkový predaj v €}
def vypocitaj_predaj_podla_produkta(objednavky):
    predaj_produktov = {}

    for objednavka in objednavky:
        produkt = objednavka["Produkt"]
        suma = objednavka["Cena"]

        if produkt in predaj_produktov:
            predaj_produktov[produkt] += suma
        else:
            predaj_produktov[produkt] = suma
    
    return(predaj_produktov)

vysledok = vypocitaj_predaj_podla_produkta(objednavky)


# 5. Funkcia: najdi_najmenej_predavany(predaj_produktov)
#    - Vstup: slovník z bodu 4
#    - Vráti produkt s najnižším predajom v € + túto hodnotu

def najdi_najmenej_predavany(predaj_produktov):
    min_predaj = float("inf")
    min_produkt = None

    for produkt in predaj_produktov:
        
        if predaj_produktov[produkt] < min_predaj:
            min_predaj = predaj_produktov[produkt]
            min_produkt = produkt

    return(min_predaj, min_produkt)

predaje = vypocitaj_predaj_podla_produkta(objednavky)

najslabsi = najdi_najmenej_predavany(predaje)


print("Súčet objednávok podľa zákazníka:")
print(vypocitaj_sucet_podla_mena(objednavky))

print("\nPočet predaných produktov:")
print(spocitaj_produkty(objednavky))

print("\nNajviac utratil:")
print(najdi_top_klienta(vypocitaj_sucet_podla_mena(objednavky)))

print("\nPredaj podľa produktov:")
print(vypocitaj_predaj_podla_produkta(objednavky))

print("\nNajmenej predávaný produkt:")
print(najdi_najmenej_predavany(vypocitaj_predaj_podla_produkta(objednavky)))
```
