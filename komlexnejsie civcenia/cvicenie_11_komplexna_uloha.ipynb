```python
# ----------------------------------------------------------------------
# KOMPLEXN√Å √öLOHA: ANAL√ùZA TR≈ΩIEB Z OBLAST√ç
# ----------------------------------------------------------------------
# M√°≈° zoznam tr≈æieb z r√¥znych oblast√≠ (mest√° a krajiny).
# Ka≈æd√Ω z√°znam obsahuje:
# - n√°zov mesta
# - krajinu
# - poƒçet predan√Ωch kusov
# - cenu za kus
#
# V zozname m√¥≈æu by≈• viacer√© z√°znamy pre rovnak√© mesto alebo krajinu.
#
# ----------------------------------------------------------------------
# ZOZNAM TR≈ΩIEB:
trzby = [
    {"Mesto": "Bratislava", "Krajina": "Slovensko", "Kusy": 20, "Cena": 9.99},
    {"Mesto": "Praha", "Krajina": "ƒåesko", "Kusy": 25, "Cena": 8.5},
    {"Mesto": "Ko≈°ice", "Krajina": "Slovensko", "Kusy": 15, "Cena": 9.99},
    {"Mesto": "Brno", "Krajina": "ƒåesko", "Kusy": 10, "Cena": 8.5},
    {"Mesto": "Ostrava", "Krajina": "ƒåesko", "Kusy": 18, "Cena": 8.5},
    {"Mesto": "Pre≈°ov", "Krajina": "Slovensko", "Kusy": 12, "Cena": 9.99}
]
#
# ----------------------------------------------------------------------
# Tvoje √∫lohy (v jednom skripte):
# 1. Vypoƒç√≠taj celkov√∫ tr≈æbu pre ka≈æd√∫ krajinu
#    ‚Üí krajina ‚Üí s√∫ƒçet tr≈æby (kusy √ó cena)
#
# 2. Zisti, ktor√° krajina zarobila najviac
#
# 3. Spoƒç√≠taj celkov√Ω poƒçet predan√Ωch kusov pre ka≈æd√© mesto
#    ‚Üí mesto ‚Üí poƒçet kusov
#
# 4. N√°jdi mesto s najv√§ƒç≈°√≠m poƒçtom predan√Ωch kusov
# ----------------------------------------------------------------------
# V≈°etko sprav najprv ako jeden veƒæk√Ω skript (bez funkci√≠).
# Potom si spoloƒçne urob√≠me refaktoring do 4 funkci√≠.
# ----------------------------------------------------------------------

trzby = [
    {"Mesto": "Bratislava", "Krajina": "Slovensko", "Kusy": 20, "Cena": 9.99},
    {"Mesto": "Praha", "Krajina": "ƒåesko", "Kusy": 25, "Cena": 8.5},
    {"Mesto": "Ko≈°ice", "Krajina": "Slovensko", "Kusy": 15, "Cena": 9.99},
    {"Mesto": "Brno", "Krajina": "ƒåesko", "Kusy": 10, "Cena": 8.5},
    {"Mesto": "Ostrava", "Krajina": "ƒåesko", "Kusy": 18, "Cena": 8.5},
    {"Mesto": "Pre≈°ov", "Krajina": "Slovensko", "Kusy": 12, "Cena": 9.99}
]


celkove_trzby_podla_krajiny = {} # 1.
top_krajina_suma = None # 2.
top_zarobok = 0 # 2.
pocet_objednavok_podla_mesta = {} # 3.
celkove_predane_kusy = {} # 4.
top_krajina_ks = None # 4.
top_kusy = 0 # 4.

for krajina in trzby:

    krajina_var = krajina["Krajina"]
    sucet_trzby = krajina["Kusy"] * krajina["Cena"]
    predane_kusy = krajina["Kusy"]
    mesto = krajina["Mesto"]
    
    # 1. Vypoƒç√≠taj celkov√∫ tr≈æbu pre ka≈æd√∫ krajinu
    #    ‚Üí krajina ‚Üí s√∫ƒçet tr≈æby (kusy √ó cena)
    if krajina_var in celkove_trzby_podla_krajiny:
        celkove_trzby_podla_krajiny[krajina_var] += sucet_trzby
    else:
        celkove_trzby_podla_krajiny[krajina_var] = sucet_trzby

    # 2. Zisti, ktor√° krajina zarobila najviac
    if celkove_trzby_podla_krajiny[krajina_var] > top_zarobok:
        top_zarobok = celkove_trzby_podla_krajiny[krajina_var]
        top_krajina_suma = krajina_var

    # 3. Spoƒç√≠taj celkov√Ω poƒçet predan√Ωch kusov pre ka≈æd√© mesto
    #    ‚Üí mesto ‚Üí poƒçet kusov
    if mesto in pocet_objednavok_podla_mesta:
        pocet_objednavok_podla_mesta[mesto] += 1
    else:
        pocet_objednavok_podla_mesta[mesto] = 1

    # 4. N√°jdi mesto s najv√§ƒç≈°√≠m poƒçtom predan√Ωch kusov
    if mesto in celkove_predane_kusy:
        celkove_predane_kusy[mesto] += predane_kusy
    else:
        celkove_predane_kusy[mesto] = predane_kusy

for krajina in celkove_predane_kusy:
    if celkove_predane_kusy[krajina] > top_kusy:
        top_kusy = celkove_predane_kusy[krajina]
        top_krajina_ks = krajina




print(celkove_trzby_podla_krajiny)
print(top_zarobok, top_krajina_suma)
print(pocet_objednavok_podla_mesta)
print(celkove_predane_kusy)
print(top_krajina_ks, top_kusy)
```

```python
# ----------------------------------------------------------------------
# üéØ CIEƒΩ: Rozbi≈• veƒæk√Ω skript na 4 samostatn√© a znovupou≈æiteƒæn√© funkcie
# ----------------------------------------------------------------------
# M√°≈° zoznam tr≈æieb (zoznam slovn√≠kov), kde ka≈æd√Ω z√°znam obsahuje:
# - n√°zov mesta
# - krajinu
# - poƒçet predan√Ωch kusov
# - cenu za kus
#
# Tento zoznam je zadan√Ω v premennej `trzby`, ktor√∫ bude≈° odovzd√°va≈• do funkci√≠.
#
# ----------------------------------------------------------------------
# TVOJE FUNKCIE BUD√ö:
#
# 1. Funkcia: vypocitaj_trzby_podla_krajiny(trzby)
#    - vstup: zoznam trzby
#    - v√Ωstup: slovn√≠k {krajina: s√∫ƒçet tr≈æby (kusy √ó cena)}
#
# 2. Funkcia: najdi_top_krajinu(trzby)
#    - vstup: zoznam trzby
#    - v√Ωstup: tuple (krajina, suma) ‚Üí krajina s najvy≈°≈°ou tr≈æbou
#
# 3. Funkcia: vypocitaj_kusy_podla_mesta(trzby)
#    - vstup: zoznam trzby
#    - v√Ωstup: slovn√≠k {mesto: poƒçet predan√Ωch kusov spolu}
#
# 4. Funkcia: najdi_top_mesto(trzby)
#    - vstup: zoznam trzby
#    - v√Ωstup: tuple (mesto, poƒçet kusov) ‚Üí mesto s najvy≈°≈°√≠m poƒçtom predan√Ωch kusov
#
# ----------------------------------------------------------------------
# POZOR:
# - ≈æiadna funkcia nesmie pou≈æ√≠va≈• glob√°lne premenn√© (v≈°etko mus√≠ by≈• vo vn√∫tri)
# - ka≈æd√° funkcia m√° jeden jasn√Ω v√Ωstup
# - keƒè bud√∫ hotov√© v≈°etky, bude≈° ich vola≈• na konci v hlavnom skripte
# ----------------------------------------------------------------------
trzby = [
    {"Mesto": "Bratislava", "Krajina": "Slovensko", "Kusy": 20, "Cena": 9.99},
    {"Mesto": "Praha", "Krajina": "ƒåesko", "Kusy": 25, "Cena": 8.5},
    {"Mesto": "Ko≈°ice", "Krajina": "Slovensko", "Kusy": 15, "Cena": 9.99},
    {"Mesto": "Brno", "Krajina": "ƒåesko", "Kusy": 10, "Cena": 8.5},
    {"Mesto": "Ostrava", "Krajina": "ƒåesko", "Kusy": 18, "Cena": 8.5},
    {"Mesto": "Pre≈°ov", "Krajina": "Slovensko", "Kusy": 12, "Cena": 9.99}
]

# 1. Funkcia: vypocitaj_trzby_podla_krajiny(trzby)
#    - vstup: zoznam trzby
#    - v√Ωstup: slovn√≠k {krajina: s√∫ƒçet tr≈æby (kusy √ó cena)}
def vypocitaj_trzby_podla_krajiny(trzby):
    celkove_trzby_podla_krajiny = {}
    for krajina in trzby:

        krajina_var = krajina["Krajina"]
        sucet_trzby = krajina["Kusy"] * krajina["Cena"]
        
        if krajina_var in celkove_trzby_podla_krajiny:
            celkove_trzby_podla_krajiny[krajina_var] += sucet_trzby
        else:
            celkove_trzby_podla_krajiny[krajina_var] = sucet_trzby
            
    return celkove_trzby_podla_krajiny

vysledok = vypocitaj_trzby_podla_krajiny(trzby) #d√°vame do premennej lebo tuto premenn√∫ m√¥≈æeme pou≈æi≈• v inej funkcii

#print(vysledok)


# 2. Funkcia: najdi_top_krajinu(trzby)
#    - vstup: zoznam trzby
#    - v√Ωstup: tuple (krajina, suma) ‚Üí krajina s najvy≈°≈°ou tr≈æbou

def najdi_top_krajinu(trzby):
 
    top_krajina_suma = None
    top_zarobok = 0

    trzby_podla_krajiny = vypocitaj_trzby_podla_krajiny(trzby)

    for krajina in trzby_podla_krajiny:

        if trzby_podla_krajiny[krajina] > top_zarobok:
            top_zarobok = trzby_podla_krajiny[krajina]
            top_krajina_suma = krajina

    return top_krajina_suma, top_zarobok

#nedal som do premennej vysledok, je to len mo≈æn√° varianta - toto ale nem√¥√Ωeme pou≈æi≈• v inej funkcii
#print(najdi_top_krajinu(trzby))

# 3. Funkcia: vypocitaj_kusy_podla_mesta(trzby)
#    - vstup: zoznam trzby
#    - v√Ωstup: slovn√≠k {mesto: poƒçet predan√Ωch kusov spolu}

def vypocitaj_kusy_podla_mesta(trzby):
    celkove_predane_kusy = {}

    for zaznam in trzby:
        predane_kusy = zaznam["Kusy"]
        mesto = zaznam["Mesto"]
        
        if mesto in celkove_predane_kusy:
            celkove_predane_kusy[mesto] += predane_kusy
        else:
            celkove_predane_kusy[mesto] = predane_kusy
    
    return celkove_predane_kusy

vysledok = vypocitaj_kusy_podla_mesta(trzby) #d√°vame do premennej lebo tuto premenn√∫ m√¥≈æeme pou≈æi≈• v inej funkcii

#print(vysledok)


# 4. Funkcia: najdi_top_mesto(trzby)
#    - vstup: zoznam trzby
#    - v√Ωstup: tuple (mesto, poƒçet kusov) ‚Üí mesto s najvy≈°≈°√≠m poƒçtom predan√Ωch kusov

def najdi_top_mesto(trzby):

    predaj = vypocitaj_kusy_podla_mesta(trzby)

    top_mesto = None
    top_kusy = 0

    for krajina in predaj:
        if predaj[krajina] > top_kusy:
            top_kusy = predaj[krajina]
            top_mesto = krajina

    return top_kusy, top_mesto

vysledok = najdi_top_mesto(trzby)

#print(vysledok)

print(vypocitaj_trzby_podla_krajiny(trzby))
print(najdi_top_krajinu(trzby))
print(vypocitaj_kusy_podla_mesta(trzby))
print(najdi_top_mesto(trzby))
```
