```python
# üßæ Zoznam objedn√°vok z e-shopu
objednavky = [
    {"produkt": "Tricko", "cena": 20, "mnozstvo": 2},
    {"produkt": "Mikina", "cena": 40, "mnozstvo": 1},
    {"produkt": "Tricko", "cena": 20, "mnozstvo": 1},
    {"produkt": "Nohavice", "cena": 50, "mnozstvo": 1},
    {"produkt": "Tricko", "cena": 20, "mnozstvo": 3},
    {"produkt": "Mikina", "cena": 40, "mnozstvo": 2},
]

# ----------------------------------------
# √öloha 1:
# Vyp√≠≈° celkov√Ω poƒçet objedn√°vok (teda poƒçet z√°znamov v zozname)
# Napr. pomocou len(objednavky)
# ----------------------------------------

# ----------------------------------------
# √öloha 2:
# Vypoƒç√≠taj celkov√Ω obrat (cena * mno≈æstvo) pre v≈°etky objedn√°vky dokopy
# (m√¥≈æe≈° pou≈æi≈• for-cyklus a medzis√∫ƒçet)
# ----------------------------------------

# ----------------------------------------
# √öloha 3:
# Zisti, koƒæko kusov ka≈æd√©ho produktu sa predalo dokopy
# Vytvor si slovn√≠k, kde key bude n√°zov produktu a value bude poƒçet kusov
# V√Ωstup:
# {'Tricko': 6, 'Mikina': 3, 'Nohavice': 1}
# ----------------------------------------

# ----------------------------------------
# √öloha 4:
# Zisti, ktor√Ω produkt mal najv√§ƒç≈°√≠ poƒçet predan√Ωch kusov (najpred√°vanej≈°√≠)
# Pou≈æi max() so slovn√≠kom z predch√°dzaj√∫cej √∫lohy
# ----------------------------------------

# ----------------------------------------
# √öloha 5:
# V≈°etky v√Ωpoƒçty zabaƒæ do funkci√≠:
# def vypocitaj_obrat(objednavky): ...
# def ziskaj_predaje_podla_produkta(objednavky): ...
# def najpredavanejsi_produkt(predaje_dict): ...
# ----------------------------------------

objednavky = [
    {"produkt": "Tricko", "cena": 20, "mnozstvo": 2},
    {"produkt": "Mikina", "cena": 40, "mnozstvo": 1},
    {"produkt": "Tricko", "cena": 20, "mnozstvo": 1},
    {"produkt": "Nohavice", "cena": 50, "mnozstvo": 1},
    {"produkt": "Tricko", "cena": 20, "mnozstvo": 3},
    {"produkt": "Mikina", "cena": 40, "mnozstvo": 2},
]


pocet_objednavok = 0 # 1.

obrat_po_produktoch = {} # 2.
celkovy_obrat = 0 # 2.

objednan√©_mnozstvo = {} # 3.

max_predan√Ωch_kusov = 0 # 4.
produkt_max_predan√Ωch_kusov = None # 4.


# √öloha 1:
# Vyp√≠≈° celkov√Ω poƒçet objedn√°vok (teda poƒçet z√°znamov v zozname)

pocet_objednavok = len(objednavky)

for objednavka in objednavky:
    produkt = objednavka["produkt"]
    mnozstvo = objednavka["mnozstvo"]
    obrat = objednavka["cena"] * objednavka["mnozstvo"]
    
    # √öloha 2:
    # Vypoƒç√≠taj celkov√Ω obrat (cena * mno≈æstvo) pre v≈°etky objedn√°vky dokopy
    if produkt in obrat_po_produktoch:
        obrat_po_produktoch[produkt] += obrat
    else:
        obrat_po_produktoch[produkt] = obrat
    celkovy_obrat += obrat

    # √öloha 3:
    # Zisti, koƒæko kusov ka≈æd√©ho produktu sa predalo dokopy

    if produkt in objednan√©_mnozstvo:
        objednan√©_mnozstvo[produkt] += mnozstvo
    else:
        objednan√©_mnozstvo[produkt] = mnozstvo
    
for produkt in objednan√©_mnozstvo:
    if objednan√©_mnozstvo[produkt] > max_predan√Ωch_kusov:
        max_predan√Ωch_kusov = objednan√©_mnozstvo[produkt]
        produkt_max_predan√Ωch_kusov = produkt


print(f"Poƒçet objedn√°vok: {pocet_objednavok}") # 1.

print(obrat_po_produktoch) # 2.
print(celkovy_obrat) # 2.

print(objednan√©_mnozstvo) # 3.

print(max_predan√Ωch_kusov, produkt_max_predan√Ωch_kusov) # 4.
```
