```python
# -----------------------------------------
# üß™ Zadanie: Sprav anal√Ωzu skladov√Ωch z√°sob
# -----------------------------------------

produkty = [
    {"nazov": "Mydlo", "cena": 2.5, "ks": 30},
    {"nazov": "≈†amp√≥n", "cena": 5.0, "ks": 15},
    {"nazov": "Zubn√° pasta", "cena": 3.2, "ks": 25},
]

# -----------------------------------------
# üîπ Funkcia 1: vypocitaj_celkovu_hodnotu(produkty)
# Vr√°≈• hodnotu z√°sob (cena * ks pre ka≈æd√Ω produkt)
# Napr. Mydlo: 2.5 * 30 = 75 ... atƒè.
# V√Ωstup: ƒç√≠slo ‚Äì celkov√° hodnota z√°sob v sklade
# -----------------------------------------


# -----------------------------------------
# üîπ Funkcia 2: najvacsi_pocet_ks(produkty)
# Vr√°≈• n√°zov produktu, ktor√Ω m√° najviac kusov na sklade
# V√Ωstup: napr. "Mydlo"
# -----------------------------------------


# -----------------------------------------
# üîπ Funkcia 3: zobraz_vysledky(hodnota, najprodukt)
# Vyp√≠≈° hodnotu z√°sob a n√°zov najviac skladovan√©ho produktu
# -----------------------------------------


# -----------------------------------------
# üîπ Funkcia 4: main()
# Zavolaj v≈°etky funkcie a spoj ich logicky
# -----------------------------------------


# -----------------------------------------
# üîπ Spustenie hlavn√©ho programu
# -----------------------------------------
# if __name__ == "__main__":
#     main()
```

```python
# üîπ Funkcia 1: vypocitaj_celkovu_hodnotu(produkty)
def vypocitaj_celkovu_hodnotu(produkty):

    predaj_produktov = {}
    celkovy_predaj = 0

    for produkt in produkty:
        nazov_produktu = produkt["nazov"]
        hodnota = produkt["cena"] * produkt["ks"]

        if nazov_produktu in predaj_produktov:
            predaj_produktov[nazov_produktu] += hodnota
        else:
            predaj_produktov[nazov_produktu] = hodnota
            
    for produkt in predaj_produktov:

        if produkt in predaj_produktov:
            celkovy_predaj += predaj_produktov[produkt]

    return celkovy_predaj

# üîπ Funkcia 2: najvacsi_pocet_ks(produkty)
def najvacsi_pocet_ks(produkty):
    naj_produkt = None
    max_pocet = 0

    for produkt in produkty:
        pocet_skladom = produkt["ks"]

        if pocet_skladom > max_pocet:
            naj_produkt = produkt["nazov"]
            max_pocet = pocet_skladom
    
    return max_pocet, naj_produkt

# üîπ Funkcia 3: zobraz_vysledky(hodnota, najprodukt)

def zobraz_vysledky(hodnota, najprodukt, max_ks):
    print(f"Produkt, ktor√©ho m√°me najviac skladom je: {najprodukt} v poƒçte {max_ks}ks.")
    print(f"Hodnota v≈°etk√Ωch produktov na sklade je: {hodnota}‚Ç¨.")


def main():
    produkty = [
    {"nazov": "Mydlo", "cena": 2.5, "ks": 30},
    {"nazov": "≈†amp√≥n", "cena": 5.0, "ks": 15},
    {"nazov": "Zubn√° pasta", "cena": 3.2, "ks": 25},
    ]

    max_ks, najprodukt = najvacsi_pocet_ks(produkty) #v tejto funkcii vraciame 2 premenne tak m√¥≈æeme 2 premenne ma≈• aj tu pre lep≈°√≠ z√°pis
    hodnota = vypocitaj_celkovu_hodnotu(produkty)

    zobraz_vysledky(hodnota, najprodukt, max_ks)

if __name__ == "__main__":
    main()
```
