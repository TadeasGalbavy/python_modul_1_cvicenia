```python
objednavky = []

with open("objednavky.csv", "r", encoding="utf-8") as file:
    riadky = file.read().splitlines() #načítanie každého riadka ako string

    hlavicka = riadky[0].split(",") #prcá riadok -> názvy stĺpcov
    data = riadky[1:] #zvyšok ->dáta

    for riadok in data:
        hodnoty = riadok.split(",")

        objednavka = {
            hlavicka[0]: hodnoty[0],
            hlavicka[1]: hodnoty[1],
            hlavicka[2]: int(hodnoty[2]),
            hlavicka[3]: float(hodnoty[3]),
        }

        objednavky.append(objednavka)

# vypocitanie súctu vsetkych objednávok a vytvorenie slovníka
sucet_objednavok = {} #zákazníci a súčet ich objednávok
zakaznici_a_sumy_obj = {} #zakazníci a sumy všetkých objednávok
max_objednávka = 0 #aktuálna najvyššia obj
najdrahsia_objednavka = None

for obj in objednavky:
    meno = obj["Meno"]
    suma_obj = obj["Pocet"] * obj["Cena"]

    # naplname dict sucet_objednavok {} kde je meno a sucet objednávok
    if meno in sucet_objednavok:
        sucet_objednavok[meno] +=(suma_obj)
    else:
        sucet_objednavok[meno] = suma_obj

    # naplname dict zakaznici_a_sumy_obj {} kde je meno a sumy objednavok
    if meno in zakaznici_a_sumy_obj:
        zakaznici_a_sumy_obj[meno].append(suma_obj)
    else:
        zakaznici_a_sumy_obj[meno] = [suma_obj]

    # naplname max_objednávka s max sumou obj
    # naplname najdrahsia_objednavka objektom, ktorý má najvyššiu sumu objednávky
    if suma_obj > max_objednávka:
        max_objednávka = suma_obj
        najdrahsia_objednavka = obj

max_suma_objednavok = 0
top_zakaznik = None


#for cyklom musíme prechádzať sucet_objednavok pretože zam máme mená a celkové súčty objednavok
for meno in sucet_objednavok: # meno = kľúče v dict
    suma = sucet_objednavok[meno] # suma = hodnoty kľúčov v premennej meno

    # hľadáme najvyššiu sumu objednávok
    if suma > max_suma_objednavok:
        max_suma_objednavok = suma
        top_zakaznik = meno

pocet_objednavok = {} #dict kde vložíme meno + súčet objednávok (vyskytov)
max_pocet_obj = 0 #definujeme si počiatočný stav počtu objednávok
max_pocet_obj_klient = None #definujeme si kľúč ktoreho hdonota bude najvyššia

#ideme prechádzať list objednavky
for klient in objednavky:
    meno = klient["Meno"] #definujeme premennú meno kam sa bude prenášať kľúč

    # ak sa meno nachádza v dict pocet_objednavok tak mu buď pripočítame jednotku alebo mu počiatočný
    # stav nastavíme na 1
    if meno in pocet_objednavok:
        pocet_objednavok[meno] += 1
    else:
        pocet_objednavok[meno] = 1

#ideme prechádzať dict pocet_objednavok
for klient in pocet_objednavok:
    pocet_obj = pocet_objednavok[klient] # do premennej si dávam hodnoty kluca klient

    if pocet_obj > max_pocet_obj:
        max_pocet_obj = pocet_obj
        max_pocet_obj_klient = klient

print("Celý zoznam klientov")
for objednavka in objednavky:
    print(objednavka)

print(f"\nZákazníci a súčet súm ich objednávok: {sucet_objednavok}")

print(f"\nZákazníci a sumy jednotlivých objednávok: {zakaznici_a_sumy_obj}")

print(f"\nAktuálne najvyššia suma objednávky je: {max_objednávka}\na patrí: {najdrahsia_objednavka["Meno"]}.")

print(f"\nObjednávky v najvyššej hodnote: {max_suma_objednavok}\nspravil {top_zakaznik}.")

print(f"\nNajvyšší počet objednávok: {max_pocet_obj}\nurobil {max_pocet_obj_klient}.")
```
